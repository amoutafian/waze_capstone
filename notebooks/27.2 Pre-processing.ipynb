```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

Goal: Create a validation set and standardize all features.

```python
#import most recent form of dataset
waze_cleaned = pd.read_csv("../data/waze_cleaned.csv")
waze_cleaned.head()
```

```python
#Create dummy or indicator features for categorical variables
dfo = waze_cleaned['device']
dfo_dummies = pd.get_dummies(dfo)
dfo_dummies.head(3)
```

```python
#recombine object features(now dummies) and numeric type features
waze_cleaned = pd.concat([dfo_dummies, waze_cleaned], axis = 1)
waze_cleaned = waze_cleaned.drop(['device'], axis = 1)
waze_cleaned.head(3)
```

```python
#Split data into testing and training datasets
X = waze_cleaned.drop('label', axis = 1)
y = waze_cleaned['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
# Standardize the magnitude of numeric features using a scaler
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
waze_cleaned.to_csv("../data/waze_processed.csv", index=False)
```
